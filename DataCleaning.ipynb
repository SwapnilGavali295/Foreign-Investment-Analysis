```python
import pandas as pd
```

```python
data=pd.read_csv('FDI data.csv')
```

```python
data
```

```python
# Define the categorization function
def categorize_sector(sector):
    sector=sector.title()
    # Heavy Industries
    heavy_industries = [
        "Metallurgical Industries",
        "Mining",
        "Power",
        "Coal Production",
        "Petroleum & Natural Gas"
    ]
    
    # Engineering and Machinery
    engineering_and_machinery = [
        "Non-Conventional Energy",
        "Industrial Instruments",
        "Automobile Industry",
        "Commercial, Office & Household Equipments",
        "Boilers And Steam Generating Plants",
        "Prime Mover (Other Than Electrical Generators)",
        "Electrical Equipments",
        "Industrial Machinery",
        "Machine Tools",
        "Agricultural Machinery",
        "Earth-Moving Machinery",
        "Miscellaneous Mechanical & Engineering Industries"
    ]
    
    # Chemical and Allied Industries
    chemical_and_allied_industries = [
        "Fertilizers",
        "Chemicals (Other Than Fertilizers)",
        "Dye-Stuffs",
        "Drugs & Pharmaceuticals",
        "Photographic Raw Film And Paper",
        "Paper And Pulp (Including Paper Products)",
        "Soaps, Cosmetics & Toilet Preparations",
        "Glue And Gelatin"
    ]
    
    # Textiles and Apparel
    textiles_and_apparel = [
        "Textiles (Including Dyed,Printed)",
        "Rubber Goods",
        "Leather,Leather Goods And Pickers"
    ]
    
    # Electronics and Technology
    electronics_and_technology = [
        "Computer Software & Hardware",
        "Electronics",
        "Telecommunications",
        "Scientific Instruments",
        "Mathematical,Surveying And Drawing Instruments"
    ]
    
    # Construction Materials
    construction_materials = [
        "Cement And Gypsum Products",
        "Timber Products",
        "Glass",
        "Ceramics"
    ]
    
    # Defense Industries
    defense_industries = [
        "Defence Industries"
    ]
    
    # Transportation and Logistics
    transportation_and_logistics = [
       
        "Air Transport (Including Air Freight)",
        "Sea Transport",
        "Ports",
        "Railway Related Components",
        "Courier Services"
    ]
    
    # Consultancy and Professional Services
    consultancy_and_professional_services = [
        "Consultancy Services",
        "Services Sector (Fin.,Banking,Insurance,Non Fin/Business,Outsourcing,R&D,Courier,Tech. Testing And Analysis, Other)"
    ]
    
    # Healthcare and Wellness
    healthcare_and_wellness = [
        "Hospital & Diagnostic Centres",
        "Pharmaceutical Services",
        "Medical And Surgical Appliances",
        
    ]
    
    # Education and Training
    education_and_training = [
        "Education"
    ]
    
    # Hospitality and Tourism
    hospitality_and_tourism = [
        "Hotel & Tourism"
    ]
    
    # Retail and Trading
    retail_and_trading = [
        "Trading",
        "Retail Trading"
    ]
    
    # Agriculture and Food Processing
    agriculture_and_food_processing = [
        "Agriculture Services",
        "Food Processing Industries",
        "Tea And Coffee (Processing & Warehousing Coffee & Rubber)",
        "Sugar",
        "Fermentation Industries",
        "Vegetable Oils And Vanaspati"
    ]
    
    # Printing and Publishing
    printing_and_publishing = [
        "Printing Of Books (Including Litho Printing Industry)",
        "Information & Broadcasting (Including Print Media)"
    ]
    
    # Miscellaneous Services
    miscellaneous_services = [
        "Diamond,Gold Ornaments",
        "Coir",
        "Construction (Infrastructure) Activities",
        "Construction Development: Townships, Housing, Built-Up Infrastructure And Construction-Development Projects",
        "Miscellaneous Industries"
    ]
    
    if sector in heavy_industries:
        return "Heavy Industries"
    elif sector in engineering_and_machinery:
        return "Engineering and Machinery"
    elif sector in chemical_and_allied_industries:
        return "Chemical and Allied Industries"
    elif sector in textiles_and_apparel:
        return "Textiles and Apparel"
    elif sector in electronics_and_technology:
        return "Electronics and Technology"
    elif sector in construction_materials:
        return "Construction Materials"
    elif sector in defense_industries:
        return "Defense Industries"
    elif sector in transportation_and_logistics:
        return "Transportation and Logistics"
    elif sector in consultancy_and_professional_services:
        return "Consultancy and Professional Services"
    elif sector in healthcare_and_wellness:
        return "Healthcare and Wellness"
    elif sector in education_and_training:
        return "Education and Training"
    elif sector in hospitality_and_tourism:
        return "Hospitality and Tourism"
    elif sector in retail_and_trading:
        return "Retail and Trading"
    elif sector in agriculture_and_food_processing:
        return "Agriculture and Food Processing"
    elif sector in printing_and_publishing:
        return "Printing and Publishing"
    elif sector in miscellaneous_services:
        return "Miscellaneous Services"
    else:
        return "Miscellaneous"
```

```python
# Melt the DataFrame to convert it into the desired format
df_melted = pd.melt(data, id_vars=["Sector"], var_name="Date", value_name="Value")
```

```python
df_melted
```

```python
# Apply categorization function to the 'Sector' column
df_melted['Category'] = df_melted['Sector'].apply(categorize_sector)
df_melted['Year']=df_melted['Date'].str[:4]
```

```python
df_melted.to_excel('FDI Data Final.xlsx',index=False)
```
